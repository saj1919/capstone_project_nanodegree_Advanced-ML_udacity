```python
hour_df = hour_df.sort_values(by=['Date', 'hour'], ascending=True)
hour_df.reset_index(inplace=True)
hour_df = hour_df[hour_features]
print hour_df.shape

hour_df.tail()
```

```python
###
### To compare correct prediction of next Ups and Down data is prepared
### If next "Closing Price" value is > previous "Closing Price" value then it is marked as 1 else 0.
###

binary_output = []
prev = 0
for rowid, row in hour_df.iterrows():
    close_val = row['Close']
    if close_val > prev: 
        binary_output.append(1)
    else:
        binary_output.append(0)
    prev = close_val
hour_df['binary_close'] = binary_output
hour_df.tail()
```

```python
###
### convert series to supervised learning
###

def series_to_supervised(data, n_in=1, n_out=1):
    n_vars = 1 if type(data) is list else data.shape[1]
    temp_df = pd.DataFrame(data)
    cols, names = list(), list()
    
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(temp_df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
        
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(temp_df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    return agg
```

```python
# ensure all data is float
values = hour_df.values.astype('float32')

# frame as supervised learning
reframed = series_to_supervised(values, 3, 1)
print reframed.shape
reframed.tail()
```

```python
###
### Remove the feature columns which are not available while future prediction
###

for col in reframed.columns:
    if "(t)" in col:
        if col in ['var1(t)', 'var2(t)', 'var3(t)', 'var4(t)', 'var5(t)', 'var6(t)', 'var13(t)']:
            del reframed[col]

print reframed.shape
reframed.fillna(0, inplace=True)
reframed.tail()
```

```python
###
### normalize features using scaling
###

scaler = MinMaxScaler(feature_range=(0, 1))
values = scaler.fit_transform(reframed.values)
```

```python
###
### Function to generate Training Data 
###

print values.shape
train, test, train_X, train_y, test_X, test_y = None, None, None, None, None, None

def train_data():
    global train, test, train_X, train_y, test_X, test_y
    # split into train and test sets
    train = values[25000:52350, :]
    test = values[52350:52425, :]

    print type(values)
    # split into input and outputs
    train_X, train_y = train[:, :-1], train[:, -1]
    test_X, test_y = test[:, :-1], test[:, -1]

    # reshape input to be 3D [samples, timesteps, features]
    train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
    test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
    print train_X.shape, train_y.shape, test_X.shape, test_y.shape
```

```python
###
### Various function for different model networks
###

def get_graph1_model():
    global train, test, train_X, train_y, test_X, test_y
    
    # input layer
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # first feature extractor
    gru = GRU(512)(visible)
    bn1 = BatchNormalization()(gru)
    do1 = Dropout(0.75)(bn1)

    # second feature extractor
    lstm = LSTM(512)(visible)
    bn2 = BatchNormalization()(lstm)
    do2 = Dropout(0.75)(bn2)

    # merge feature extractors
    merge = concatenate([do1, do2])

    # interpretation layer
    hidden1 = Dense(512)(merge)
    bn4 = BatchNormalization()(hidden1)
    do4 = Dropout(0.75)(bn4)

    # prediction output
    output = Dense(1)(do4)
    model = Model(inputs=visible, outputs=output)
    return model
    
def get_graph2_model():
    global train, test, train_X, train_y, test_X, test_y
    
    model = Sequential()
    model.add(Convolution1D(input_shape = (train_X.shape[1], train_X.shape[2]),
                            nb_filter=168,
                            filter_length=32,
                            border_mode='same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU())
    model.add(Dropout(0.25))

    model.add(Convolution1D(nb_filter=84,
                            filter_length=16,
                            border_mode='same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU())
    model.add(Dropout(0.25))

    model.add(Convolution1D(nb_filter=24,
                            filter_length=8,
                            border_mode='same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU())
    model.add(Dropout(0.25))

    model.add(Convolution1D(nb_filter=8,
                            filter_length=4,
                            border_mode='same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU())
    model.add(Dropout(0.25))

    model.add(Flatten())

    model.add(Dense(512))
    model.add(BatchNormalization())
    model.add(LeakyReLU())

    model.add(Dense(512))
    model.add(BatchNormalization())
    model.add(LeakyReLU())

    model.add(Dense(1))
    return model

def get_graph3_model():
    global train, test, train_X, train_y, test_X, test_y
    
    # input layer
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # first feature extractor
    gru1 = GRU(384)(visible)
    bn1 = BatchNormalization()(gru1)
    do1 = Dropout(0.5)(bn1)

    # second feature extractor
    lstm2 = LSTM(384)(visible)
    bn2 = BatchNormalization()(lstm2)
    do2 = Dropout(0.5)(bn2)

    # Third feature extractor
    gru3 = Bidirectional(GRU(384))(visible)
    bn3 = BatchNormalization()(gru3)
    do3 = Dropout(0.5)(bn3)

    # Forth feature extractor
    lstm4 = Bidirectional(LSTM(384))(visible)
    bn4 = BatchNormalization()(lstm4)
    do4 = Dropout(0.5)(bn4)

    # merge feature extractors
    merge = concatenate([do1, do2, do3, do4])


    # interpretation layer
    hidden1 = Dense(384)(merge)
    bn5 = BatchNormalization()(hidden1)
    do5 = Dropout(0.5)(bn5)

    # prediction output
    output = Dense(1)(do5)
    model = Model(inputs=visible, outputs=output)
    return model
```

```python
def run_rnn(with_model='graph1'):
    global train, test, valid, train_X, train_y, test_X, test_y
    train_data()
    
    if with_model=='graph1':
        model = get_graph1_model()
    elif with_model=='graph2':
        model = get_graph2_model()
    elif with_model=='graph3':
        model = get_graph3_model()
    model.compile(loss='mse', optimizer='adam')
    print model.summary()

    earlystop = EarlyStopping(monitor='loss', min_delta=0.0000001, patience=96, verbose=1, mode='auto')
    checkpointer = ModelCheckpoint(filepath='../output/tmp/model_lstm.hdf5', verbose=1, save_best_only=True)
    callbacks_list = [earlystop, checkpointer]

    # fit network
    history = model.fit(train_X, train_y, epochs=512, batch_size=384, 
                        validation_data=(test_X, test_y), callbacks=callbacks_list, shuffle=False)

    # plot history
    plt.plot(history.history['loss'][1:], label='train')
    plt.plot(history.history['val_loss'][1:], label='test')
    plt.legend()
    plt.show()

    if with_model=='graph1':
        model = get_graph1_model()
    elif with_model=='graph2':
        model = get_graph2_model()
    elif with_model=='graph3':
        model = get_graph3_model()
    model.load_weights("../output/tmp/model_lstm.hdf5")
    model.compile(loss='mse', optimizer='adam')
    print model.summary()

    # make a prediction
    yhat = model.predict(test_X)

    # calculate RMSE
    rmse = np.sqrt(mean_squared_error(test_y, yhat))
    print 'Test RMSE: %.3f' % rmse

    result = pd.DataFrame()
    result['actual'] = test_y
    result['prediction'] = yhat
    result.plot()

    for limit in np.arange(0.0, 1.0, 0.05):
        print "==== ",limit, " ===="
        result['pred'] = 0
        result.loc[result['prediction']>limit, 'pred'] = 1
        print classification_report(result['actual'], result['pred'])
        print "\n\n"
    
```

```python
run_rnn('graph1')
```
