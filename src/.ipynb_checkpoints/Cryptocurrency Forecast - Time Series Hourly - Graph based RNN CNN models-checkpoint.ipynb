```python
###
### convert series to supervised learning
###

def series_to_supervised(data, n_in=1, n_out=1):
    n_vars = 1 if type(data) is list else data.shape[1]
    temp_df = pd.DataFrame(data)
    cols, names = list(), list()
    
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(temp_df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
        
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(temp_df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    return agg
```

```python
###
### Function to generate Training Data 
###

print values.shape
train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y = None, None, None, None, None, None, None, None, None
def gen_train_data():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    # split into train and test sets
    train = values[25000:52350, :]
    test = values[52350:52450, :]
    valid = values[np.random.choice(values[40000:].shape[0], 5000, replace=False), :]

    print type(values)
    # split into input and outputs
    train_X, train_y = train[:, :-1], train[:, -1]
    test_X, test_y = test[:, :-1], test[:, -1]
    valid_X, valid_y = valid[:, :-1], valid[:, -1]

    # reshape input to be 3D [samples, timesteps, features]
    train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
    test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
    valid_X = valid_X.reshape((valid_X.shape[0], 1, valid_X.shape[1]))
    print train_X.shape, train_y.shape, test_X.shape, test_y.shape, valid_X.shape, valid_y.shape
```

```python
###
### Various function for different model networks
###

def get_graph1_model():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    # input layer
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # first feature extractor
    gru = LSTM(256)(visible)
    bn1 = BatchNormalization()(gru)
    do1 = Dropout(0.5)(bn1)

    # second feature extractor
    lstm = LSTM(512)(visible)
    bn2 = BatchNormalization()(lstm)
    do2 = Dropout(0.5)(bn2)

    # merge feature extractors
    merge = concatenate([do1, do2])

    # interpretation layer
    hidden1 = Dense(384)(merge)
    bn4 = BatchNormalization()(hidden1)
    do4 = Dropout(0.5)(bn4)

    # prediction output
    output = Dense(1)(do4)
    model = Model(inputs=visible, outputs=output)

    return model

def get_graph2_model():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    # input layer
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # first feature extractor
    gru = GRU(256)(visible)
    bn1 = BatchNormalization()(gru)
    do1 = Dropout(0.5)(bn1)

    # second feature extractor
    lstm = GRU(512)(visible)
    bn2 = BatchNormalization()(lstm)
    do2 = Dropout(0.5)(bn2)

    # merge feature extractors
    merge = concatenate([do1, do2])

    # interpretation layer
    hidden1 = Dense(384)(merge)
    bn4 = BatchNormalization()(hidden1)
    do4 = Dropout(0.5)(bn4)

    # prediction output
    output = Dense(1)(do4)
    model = Model(inputs=visible, outputs=output)

    return model

def get_graph3_model():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    # input layer
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # first feature extractor
    gru = GRU(384)(visible)
    bn1 = BatchNormalization()(gru)
    do1 = Dropout(0.5)(bn1)

    # second feature extractor
    lstm = LSTM(384)(visible)
    bn2 = BatchNormalization()(lstm)
    do2 = Dropout(0.5)(bn2)

    # merge feature extractors
    merge = concatenate([do1, do2])

    # interpretation layer
    hidden1 = Dense(384)(merge)
    bn4 = BatchNormalization()(hidden1)
    do4 = Dropout(0.5)(bn4)

    # prediction output
    output = Dense(1)(do4)
    model = Model(inputs=visible, outputs=output)

    return model

def get_graph4_model():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    # input layer
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # first feature extractor
    gru = GRU(384)(visible)
    bn1 = BatchNormalization()(gru)
    do1 = Dropout(0.5)(bn1)

    # second feature extractor
    lstm = LSTM(384)(visible)
    bn2 = BatchNormalization()(lstm)
    do2 = Dropout(0.5)(bn2)

    # merge feature extractors
    merge = concatenate([do1, do2])

    # interpretation layer
    hidden1 = Dense(384)(merge)
    bn4 = BatchNormalization()(hidden1)
    do4 = Dropout(0.5)(bn4)
    
    # interpretation layer
    hidden2 = Dense(384)(do4)
    bn5 = BatchNormalization()(hidden2)
    do5 = Dropout(0.5)(bn5)

    # prediction output
    output = Dense(1)(do4)
    model = Model(inputs=visible, outputs=output)

    return model


def get_graph5_model():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # cnn extractor
    cnn11 = Convolution1D(nb_filter=32, filter_length=8, border_mode='same')(visible)
    bn11 = BatchNormalization()(cnn11)
    lrelu11 = LeakyReLU()(bn11)
    do11 = Dropout(0.5)(lrelu11)
    
    f14 = Flatten()(do11)
    den14 = Dense(384)(f14)
    bn14 = BatchNormalization()(den14)
    lrelu14 = LeakyReLU()(bn14)
    do14 = Dropout(0.5)(lrelu14)
    
    # prediction output
    output = Dense(1)(do14)
    model = Model(inputs=visible, outputs=output)
    return model

def get_graph6_model():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # cnn extractor 1
    cnn11 = Convolution1D(nb_filter=32, filter_length=8, border_mode='same')(visible)
    bn11 = BatchNormalization()(cnn11)
    lrelu11 = LeakyReLU()(bn11)
    do11 = Dropout(0.5)(lrelu11)
    
    cnn12 = Convolution1D(nb_filter=16, filter_length=4, border_mode='same')(do11)
    bn12 = BatchNormalization()(cnn12)
    lrelu12 = LeakyReLU()(bn12)
    do12 = Dropout(0.5)(lrelu12)
    
    f14 = Flatten()(do12)
    den14 = Dense(384)(f14)
    bn14 = BatchNormalization()(den14)
    lrelu14 = LeakyReLU()(bn14)
    do14 = Dropout(0.5)(lrelu14)
    
    # prediction output
    output = Dense(1)(do14)
    model = Model(inputs=visible, outputs=output)
    return model

def get_graph7_model():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # cnn extractor 1
    cnn11 = Convolution1D(nb_filter=32, filter_length=8, border_mode='same')(visible)
    bn11 = BatchNormalization()(cnn11)
    lrelu11 = LeakyReLU()(bn11)
    do11 = Dropout(0.5)(lrelu11)
    
    cnn12 = Convolution1D(nb_filter=16, filter_length=4, border_mode='same')(do11)
    bn12 = BatchNormalization()(cnn12)
    lrelu12 = LeakyReLU()(bn12)
    do12 = Dropout(0.5)(lrelu12)
    
    cnn13 = Convolution1D(nb_filter=8, filter_length=4, border_mode='same')(do12)
    bn13 = BatchNormalization()(cnn13)
    lrelu13 = LeakyReLU()(bn13)
    do13 = Dropout(0.5)(lrelu13)
    
    f14 = Flatten()(do13)
    den14 = Dense(384)(f14)
    bn14 = BatchNormalization()(den14)
    lrelu14 = LeakyReLU()(bn14)
    do14 = Dropout(0.5)(lrelu14)
    
    # prediction output
    output = Dense(1)(do14)
    model = Model(inputs=visible, outputs=output)
    return model

def get_graph8_model():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    # input layer
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # first feature extractor
    gru = Bidirectional(GRU(384))(visible)
    bn1 = BatchNormalization()(gru)
    do1 = Dropout(0.5)(bn1)

    # second feature extractor
    lstm = Bidirectional(LSTM(384))(visible)
    bn2 = BatchNormalization()(lstm)
    do2 = Dropout(0.5)(bn2)

    # merge feature extractors
    merge = concatenate([do1, do2])

    # interpretation layer
    hidden1 = Dense(384)(merge)
    bn4 = BatchNormalization()(hidden1)
    do4 = Dropout(0.5)(bn4)

    # prediction output
    output = Dense(1)(do4)
    model = Model(inputs=visible, outputs=output)

    return model


def get_graph9_model():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    # input layer
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # first feature extractor
    gru1 = GRU(384)(visible)
    bn1 = BatchNormalization()(gru1)
    do1 = Dropout(0.5)(bn1)

    # second feature extractor
    lstm2 = LSTM(384)(visible)
    bn2 = BatchNormalization()(lstm2)
    do2 = Dropout(0.5)(bn2)
    
    # first feature extractor
    gru3 = Bidirectional(GRU(384))(visible)
    bn3 = BatchNormalization()(gru3)
    do3 = Dropout(0.5)(bn3)

    # second feature extractor
    lstm4 = Bidirectional(LSTM(384))(visible)
    bn4 = BatchNormalization()(lstm4)
    do4 = Dropout(0.5)(bn4)

    # merge feature extractors
    merge = concatenate([do1, do2, do3, do4])

    # interpretation layer
    hidden5 = Dense(384)(merge)
    bn5 = BatchNormalization()(hidden5)
    do5 = Dropout(0.5)(bn5)

    # prediction output
    output = Dense(1)(do5)
    model = Model(inputs=visible, outputs=output)

    return model


def get_graph10_model():
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y
    # input layer
    visible = Input(shape=(train_X.shape[1],train_X.shape[2]))

    # first feature extractor
    gru1 = GRU(384)(visible)
    bn1 = BatchNormalization()(gru1)
    do1 = Dropout(0.5)(bn1)

    # second feature extractor
    lstm2 = LSTM(384)(visible)
    bn2 = BatchNormalization()(lstm2)
    do2 = Dropout(0.5)(bn2)
    
    # third feature extractor
    gru3 = Bidirectional(GRU(384))(visible)
    bn3 = BatchNormalization()(gru3)
    do3 = Dropout(0.5)(bn3)

    # forth feature extractor
    lstm4 = Bidirectional(LSTM(384))(visible)
    bn4 = BatchNormalization()(lstm4)
    do4 = Dropout(0.5)(bn4)
    
    # cnn extractor
    cnn11 = Convolution1D(nb_filter=32, filter_length=8, border_mode='same')(visible)
    bn11 = BatchNormalization()(cnn11)
    lrelu11 = LeakyReLU()(bn11)
    do11 = Dropout(0.5)(lrelu11)
    cnn12 = Convolution1D(nb_filter=16, filter_length=4, border_mode='same')(do11)
    bn12 = BatchNormalization()(cnn12)
    lrelu12 = LeakyReLU()(bn12)
    do12 = Dropout(0.5)(lrelu12)
    cnn13 = Convolution1D(nb_filter=8, filter_length=4, border_mode='same')(do12)
    bn13 = BatchNormalization()(cnn13)
    lrelu13 = LeakyReLU()(bn13)
    do13 = Dropout(0.5)(lrelu13)
    f13 = Flatten()(do13)
    den13 = Dense(384)(f13)
    bn13 = BatchNormalization()(den13)
    lrelu13 = LeakyReLU()(bn13)
    do0 = Dropout(0.5)(lrelu13)

    # merge feature extractors
    merge = concatenate([do0, do1, do2, do3, do4])

    # interpretation layer
    hidden5 = Dense(384)(merge)
    bn5 = BatchNormalization()(hidden5)
    do5 = Dropout(0.5)(bn5)

    # prediction output
    output = Dense(1)(do5)
    model = Model(inputs=visible, outputs=output)

    return model
```

```python
def run_rnn(with_model='graph1'):
    global train, test, valid, train_X, train_y, test_X, test_y, valid_X, valid_y 
    gen_train_data()
    
    if with_model=='graph1':
        model = get_graph1_model()
    elif with_model=='graph2':
        model = get_graph2_model()
    elif with_model=='graph3':
        model = get_graph3_model()
    elif with_model=='graph4':
        model = get_graph4_model()
    elif with_model=='graph5':
        model = get_graph5_model()
    elif with_model=='graph6':
        model = get_graph6_model()
    elif with_model=='graph7':
        model = get_graph7_model()
    elif with_model=='graph8':
        model = get_graph8_model()
    elif with_model=='graph9':
        model = get_graph9_model()
    elif with_model=='graph10':
        model = get_graph10_model()
        
    model.compile(loss='mse', optimizer='adam')
    print model.summary()

    earlystop = EarlyStopping(monitor='loss', min_delta=0.0000001, patience=64, verbose=1, mode='auto')
    checkpointer = ModelCheckpoint(filepath='../output/tmp/model_lstm.hdf5', verbose=1, save_best_only=True)
    callbacks_list = [earlystop, checkpointer]

    # fit network
    history = model.fit(train_X, train_y, epochs=500, batch_size=384, 
                        validation_data=(test_X, test_y), callbacks=callbacks_list, shuffle=False)

    # plot history
    plt.plot(history.history['loss'][:], label='train')
    plt.plot(history.history['val_loss'][:], label='test')
    plt.legend()
    plt.show()

    if with_model=='graph1':
        model = get_graph1_model()
    elif with_model=='graph2':
        model = get_graph2_model()
    elif with_model=='graph3':
        model = get_graph3_model()
    elif with_model=='graph4':
        model = get_graph4_model()
    elif with_model=='graph5':
        model = get_graph5_model()
    elif with_model=='graph6':
        model = get_graph6_model()
    elif with_model=='graph7':
        model = get_graph7_model()
    elif with_model=='graph8':
        model = get_graph8_model()
    elif with_model=='graph9':
        model = get_graph9_model()
    elif with_model=='graph10':
        model = get_graph10_model()
    model.load_weights("../output/tmp/model_lstm.hdf5")
    model.compile(loss='mse', optimizer='adam')
    print model.summary()


    # make a prediction
    yhat = model.predict(test_X)
    test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))

    # invert scaling for forecast
    inv_yhat = np.concatenate((test_X, yhat), axis=1)
    inv_yhat = scaler.inverse_transform(inv_yhat)
    inv_yhat = inv_yhat[:,-1]

    # invert scaling for actual
    test_y = test_y.reshape((len(test_y), 1))
    inv_y = np.concatenate((test_X, test_y), axis=1)
    inv_y = scaler.inverse_transform(inv_y)
    inv_y = inv_y[:,-1]

    # calculate RMSE
    rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat))
    print 'Test RMSE: %.3f' % rmse
    result = pd.DataFrame()
    result['actual'] = inv_y
    result['prediction'] = inv_yhat
    result.plot()

    ###
    ### To compare correct prediction of next Ups and Down data is prepared
    ### If next "Closing Price" value is > previous "Closing Price" value then it is marked as 1 else 0.
    ###
    binary_op_actual, binary_op_pred = [], []
    prev_val = 0
    for i, val in enumerate(result['actual'].values):
        if i > 0:
            if val > prev_val:
                binary_op_actual.append(1)
            else:
                binary_op_actual.append(0)
        prev_val = val

    prev_val = 0
    for i, val in enumerate(result['prediction'].values):
        if i > 0:
            if val > prev_val:
                binary_op_pred.append(1)
            else:
                binary_op_pred.append(0)
        prev_val = val

    print len(binary_op_actual), len(binary_op_pred)
    print binary_op_actual[:10], binary_op_pred[:10]

    print classification_report(binary_op_actual, binary_op_pred)
```
