```python
import pandas as pd
import pandas_datareader.data as web
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import os

plt.style.use('fivethirtyeight')
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

## Data generation Using Yahoo pandas web api
Returns Daily trend 

```python
df = None
for crypto in ['BTC', 'LTC', 'ETH', 'XRP', 'BCH', 'ADA', 'XLM', 'EOS', 'XEM', 'NEO', 'IOT', 'DASH', 'XMR', 'TRX']:
    print "Adding Data for %s" % crypto
    temp_df = web.get_data_yahoo('%s-USD'%crypto, start=datetime.datetime(2015, 1, 1), end=datetime.datetime(2018, 1, 28))
    temp_df.reset_index(inplace=True)
    del temp_df['Adj Close']
    temp_df.columns = ['Date'] + [x+"_%s"%crypto for x in temp_df.columns if x != 'Date']
    if df is None:
        df = temp_df.copy()
    else:
        df = pd.merge(df, temp_df, on='Date', how='left')
print df.columns
df.head()
```

## Data preparation of different crypto features
Data is downloaded from blockchain.com and prepared daily price trend dataframe

```python
df['Date'] = pd.to_datetime(df['Date'])

for f in os.listdir("../data/blockchain_data/"):
    if f.endswith(".csv"):
        feature = str(f).replace(".csv", "")
        print "Adding ", feature
        temp_df = pd.read_csv('../data/blockchain_data/%s' % f, header=-1)
        temp_df.columns = ['Date', feature]
        temp_df['Date'] = pd.to_datetime(temp_df['Date'])

        df = pd.merge(df, temp_df, on='Date', how='left')
print df.shape
df.head()
```

## Replace None values with zero

```python
df.fillna(0, inplace=True)
df.tail()
```

## Save DataFrame for later use

```python
df.to_csv('../output/BTC_Crypto_Data.csv', index=False, sep='\t')
```

## Minute-by-Minute Data From Kaggle.com

```python
df = pd.read_csv('../data/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2018-01-08.csv', sep=',', header=0)
```

```python
def timestamp_to_datetime_conversion(row):
    return datetime.datetime.fromtimestamp(row['Timestamp']).strftime('%Y-%m-%d %H:%M:%S')
        
df['datetime'] = df.apply(timestamp_to_datetime_conversion, axis=1)
```

```python
df['datetime_pandas'] = pd.to_datetime(df['datetime'])
df['day_of_month'] = df['datetime_pandas'].dt.day
df['month'] = df['datetime_pandas'].dt.month
df['year'] = df['datetime_pandas'].dt.year
df['weekday'] = df['datetime_pandas'].dt.weekday
df['week'] = df['datetime_pandas'].dt.week
df['hour'] = df['datetime_pandas'].dt.hour
df['minute'] = df['datetime_pandas'].dt.minute
```

## Save DataFrame for later use

```python
df.to_csv('../data/bitstamp_minute_data.csv', sep=',', index=False)
```
